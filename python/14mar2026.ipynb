```python
ii=101
i=ii
a=0
while i>0:
    # by dividng a number by 10, the reminder will be the last digit
    last=i%10
    
    a=(a*10)+last
    i//=10
if ii==a:
    print('palindrome')
else:
    print('not palindrome')
```

```python
ii=101
i=ii
a=0
while i>0:
    # by dividng a number by 10, the reminder will be the last digit
    last=i%10
    #here as a is updating, if a will be added with last instead of keeping in position
    a+=last
    print(a)
    i//=10
if ii==a:
    print('palindrome')
else:
    print('not palindrome')
```
