```python
def clean(text):
    texts=''
    for ch in text:
        if ch.isalnum() or ch==' ':
            texts+=ch.lower()
    return texts
```

```python
text_in=input('enter the web text')
cleaned_text=clean(text_in)
```

```python
print(cleaned_text)
```

```python
def ch_token(text):
    clean_text=clean(text)
    ch_text=[]
    for ch in clean_text:
        if ch!=' ':
            ch_text.append(ch)
        else:
            ch_text.append('</w>')
    return ch_text
    
```

```python
text_in=input('enter the web text')
cleaned_text=ch_token(text_in)
```

```python
print(cleaned_text)
```

```python
def count_pair(text):
    ch_tokens=ch_token(text)
    count_pair={}
    for i in range(0,len(ch_tokens)-1):
        if ch_tokens[i]=='</w>':
            continue
        pair=(ch_tokens[i]+ch_tokens[i+1])
        if '</w>' not in pair:
            if pair not in count_pair:
                count_pair[pair]=1
            else:
                count_pair[pair]+=1
    return count_pair
```

```python
text_in=input('enter the web text')
count=count_pair(text_in)
print(count)
```
