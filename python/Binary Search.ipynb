# Binary Search
### Binary search is an optimized algorithm for finding an item from a sorted list of items.
### it start with middle: look at element in middle of sorted list.
### if same print
### if target is smaller than middle element, discard right half.
### if target is larger than middle element, discard left half.
### repeat this until the target is found.

```python
a=[1,2,3,4,5,7,23,34,45,56,67]
target=67
start=0
end=len(a)-1
mid=0
while start<=end:
    mid=(start+end)//2
    if target==a[mid]:
        print('target is in index',mid)
        break
    elif a[mid]<target:
        start=mid+1
    else:
        end=mid-1
if start>end:
    print('found in index')
```
