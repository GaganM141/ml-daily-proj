# Neon numbers
Neon number is a number where the sum of digits of its square is equal to original number.
There exist only 3 neon numbers: 0, 1, 9
ex: 9
    9**2=81
    

```python


neon=int(input('enter a number:'))
temp=neon**2
sum=0
while temp>0:
    last=temp%10
    sum+=last
    temp//=10
if sum==neon:
    print('neon')
else:
    print('not neon')
    
```
