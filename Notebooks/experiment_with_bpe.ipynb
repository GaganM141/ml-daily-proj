```python
def clean(text):
    texts=''
    for ch in text:
        if ch.isalnum() or ch==' ':
            texts+=ch.lower()
    return texts
```

```python
def ch_token(text):
    clean_text=clean(text)
    ch_text=[]
    for ch in clean_text:
        if ch!=' ':
            ch_text.append(ch)
        else:
            ch_text.append('</w>')
    #print('ch_text:',ch_text,sep='\n',end='\n\n\n')
    return ch_text
    
```

```python
def count_pair(text):
    ch_tokens=text
    Dcount_pair={}
    for i in range(0,len(ch_tokens)-1):
        if ch_tokens[i]=='</w>':
            continue
        pair=(ch_tokens[i],ch_tokens[i+1])
        if '</w>' not in pair:
            if pair not in Dcount_pair:
                Dcount_pair[pair]=1
            else:
                Dcount_pair[pair]+=1
    #print("dcount_pair",Dcount_pair,sep='\n')
    return Dcount_pair
```

```python
def best_pair(text):
    pair_count=text
    max_count=0
    best=None
    for pair in pair_count:
        #print('pair',pair,'   pair_value',pair_count[pair],sep='     ',end='\n')
        if pair_count[pair]>max_count:
            max_count=pair_count[pair]
            #print('max count    ',max_count)
            best=pair
    #print('best',best)
    return best
            
```

```python
def get_vocab(tokens):

    vocab=set()

    for tok in tokens:
        vocab.add(tok)

    return vocab
```

```python
def train_bpe(text, max_vocab):
    tokens = ch_token(text)
    merges = []
    while True:
        pair_counts = count_pair(tokens)
        pair = best_pair(pair_counts)
        if pair is None:
            break
        if pair_counts[pair] == 1:
            break
        merges.append(pair)
        tokens = merge_pair(tokens, pair)
        vocab = get_vocab(tokens)
        print("merge:", pair, " vocab size:", len(vocab))
        if len(vocab) >= max_vocab:
            break
    return tokens, merges
```

```python
def build_vocab(vocab):

    vocab_dict = {}
    id = 0

    for token in vocab:
        vocab_dict[token] = id
        id += 1

    return vocab_dict
```
