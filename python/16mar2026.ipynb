# Perfect number
a number whose sum of divisor is equal to the number
ex: 6
6 is divisable by 1,2,3
and 1+2+3=6

```python
n=6
summ=0
for i in range(1,n):
    if n%i==0:
        summ+=i
if summ==n:
    print('perfect number')
else:
    print('not a perfect number')
```

# Strong number
## a number whose sum of digit factorial is equal to the number
### ex: 145
1!+4!+5!=145

```python
n=145
temp=n
summ=0
while temp>0:
    last=temp%10
    print('last:',last)
    factorial=1
    for i in range(1,last+1):
        factorial*=i
        print('factorial:',factorial)
    summ+=factorial  
    print('sum',summ)
    temp//=10
if summ==n:
    print('strong number')
else:
    print('not a strong number')
```
