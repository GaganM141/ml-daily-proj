```python
str=input('enter a string:')
reverse_str=''
i=1
while i<=len(str):
    reverse_str+=str[-i]
    i+=1
print(reverse_str)
```

```python
str=input('enter a string:')
rev=""
for i in str:
    rev=i+rev
print(rev)
```
