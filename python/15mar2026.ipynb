```python
#Fibonacci
fib=int(input('enter a number to get fibonacci:'))
j=0
k=0
l=0
while j<=fib:
    k+=j
    print(k)
    l+=j+1
    print(l)
    j+=1
    print(end='\n\n')
print(k)
    
    
```
