# Perfect number
## a number whose sum of divisor is equal to the number
## ex: 6
## 6 is divisable by 1,2,3
## and 1+2+3=6

```python
n=6
summ=0
for i in range(1,n):
    if n%i==0:
        summ+=i
if summ==n:
    print('perfect number')
else:
    print('not a perfect number')
```

# Strong number
## a number whose sum of digit factorial is equal to the number
### ex: 145
1!+4!+5!=145

```python
n=145
temp=n
summ=0
while temp>0:
    # here we get last digit of the number.
    last=temp%10
    #factorial should not be zero as we are doing multiplication
    #factorial should be inside while loop but outside for loop because the factorial should be updated for each digit
    factorial=1
    # for loop is used to do factorial of the last number
    for i in range(1,last+1):
        factorial*=i
    # adding the factorial of last digit to summ
    print(last,'!+',end='')
    summ+=factorial  
    # updating temp number without last digit
    temp//=10
if summ==n:
    print(' \n strong number')
else:
    print(' \n not a strong number')
```

# Spy number
## a positive integer where sum of its digits is equal to product of its digits
## ex: 1124
## 1+1+2+4=8
## 1x1x2x4=8


```python
n=1124
summ=0
product=1
temp=n
while temp>0:
    last=temp%10
    summ+=last
    product*=last
    temp//=10
if product==summ:
    print('it is a spy number','\n product=',product,'\n sum=',summ)
else:
    print('it is not a spy number','\n product=',product,'\n sum=',summ)
```

# Buzz number
## a number that either ends with 7 or is divisable by 7

```python
num=21
last=num%10
if n%7==0 or last==7:
    print('the number is a buzz number:',num)
else:
    print('the number is not a buzz number:',num)
```
