```python
s='hello everyone how are you'
res={}
for i in s.split():
    f=i[0]+i[len(i)-1]
    cons,count,list1="",0,[]
    for j in i:
        if j not in 'AEIOUaeiou' and ('A'<=j<='Z' or 'a'<=j<='z'):
            count+=1
            cons+=j
    res[f]=[cons,count]
print(res)
```

# input is s='Nested loop '
# output res={'detseN':6, 'pool':4}

```python
s='Nested loop is a loop inside another loop'
res={}
for i in s.split():
    count=0
    for j in i:
        count+=1
        res[i[::-1]]=count
print(res)
    
    
```

# input is s='kabab is love'
# output res={'kabab':['babak', 2, 'kbb'], 'is':['si',1,'i'],'love':['evol',2,'lv']}
## [reverse, no of vowels, char at even index]

```python
s='kabab is love'
res={}
for i in s.split():
    count,vowl=0,''
    for j in range(len(i)):
        if i[j] in 'AEIOUaeiou':
            count+=1
        if j%2==0:
            vowl+=i[j]
    res[i]=[i[::-1],count,vowl]
print(res)
```

# input is s='kabab is love'
# output res={'kb':('kbb',3,'bbk'), 'is':('s',1,'s'), 'le':('lv',2,'vl')}
## {1st+last char: (consonant, number of consonant, rev of consonant)}

```python
s='for loop is also called as self iterating loop'
res={}
for i in s.split():
    k=i[0]+i[len(i)-1]
    cons,count='',0
    for j in i:
        if ('A'<=j<='Z' and 'a'<=j<='z') or j not in 'AEIOUaeiou':
            count+=1
            cons+=j
    res[k]=(cons,count,cons[::-1])
print(res)
```

```python
inp=[100, 200, 90, 80, 300, 10, 20]
out=[]
for i in range(len(inp)):
    num=0
    for j in range(len(inp)):
        if i==j:
            continue
        else:
            num+=inp[j]
    out.append(num)
            
print(out)
```
