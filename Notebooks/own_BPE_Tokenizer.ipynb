```python
""" function to clean the text.
look through each character and only keep valid characters.
this removes https, emoji, punctuations, etc"""
def clean_text(text):
    cleaned=""
    for ch in text:
        if ch.isalnum() or ch==" ":
            cleaned += ch.lower()
    return cleaned
```

```python
''' Splitting words at spaces '''
def split_words(text):
    words=[]
    current_word=""
    for ch in text:
        if ch!=" ":
            '''current_word+= current_word this line is not correct as it
            adds to current words
            example : current_words='hi'
                      after 
                      current_words = 'hihi'
            '''
            current_word+=ch
        else:
            if current_word:
                words.append(current_word)
                # here we reset current_word after appending a word to list words
                current_word=""
    return words
```

```python
''' converting words into character tokens'''
def word_to_tokens(word):
    tokens=[]
    for ch in word:
        tokens.append(ch)
    '''BPE needs end of word marker, which helps where the word ends
    else BPE will merge across word boundaries incorrectly.'''
    tokens.append("</w>")
    return tokens
```

```python
'''
Vocabulary frequency 
count how often each word appears
'''
def vocab(words):
    vocab={}
    for word in words:
        tokens=tuple(word_to_tokens(word))
        if tokens not in vocab:
            vocab[tokens]=1
        else:
            vocab[tokens]+=1
    return vocab
```

```python
''' 
count pair frequencies
'''
def count_pairs(vocab):
    pair_freq={}
    for token_list in vocab:
        freq= vocab[token_list]
        for i in range( 0, len(token_list)-2):
            pair=(token_list[i],token_list[i+1])
            pair_freq[pair]+=freq
    return pair_freq
```
