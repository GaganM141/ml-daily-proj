```python
def clean(text):
    texts=''
    for ch in text:
        if ch.isalnum() or ch==' ':
            texts+=ch.lower()
    return texts
```

```python
def ch_token(text):
    clean_text=clean(text)
    ch_text=[]
    token={}
    n=0
    for ch in clean_text:
        if ch!=' ':
            ch_text.append(ch)
            if ch not in token:
                token[ch]=n+1
        else:
            ch_text.append('</w>')
    #print('ch_text:',ch_text,sep='\n',end='\n\n\n')
    return ch_text
    
```

```python
def count_pair(text):
    ch_tokens=text
    Dcount_pair={}
    for i in range(0,len(ch_tokens)-1):
        if ch_tokens[i]=='</w>':
            continue
        pair=(ch_tokens[i],ch_tokens[i+1])
        if '</w>' not in pair:
            if pair not in Dcount_pair:
                Dcount_pair[pair]=1
            else:
                Dcount_pair[pair]+=1
    #print("dcount_pair",Dcount_pair,sep='\n')
    return Dcount_pair
```

```python
def best_pair(text):
    pair_count=text
    max_count=0
    best=None
    for pair in pair_count:
        #print('pair',pair,'   pair_value',pair_count[pair],sep='     ',end='\n')
        if pair_count[pair]>max_count:
            max_count=pair_count[pair]
            #print('max count    ',max_count)
            best=pair
    #print('best',best)
    return best
            
```

```python
def merge_pair(tokens, pair):

    new_tokens = []
    i = 0

    while i < len(tokens):

        if i < len(tokens)-1 and tokens[i] == pair[0] and tokens[i+1] == pair[1]:

            new_tokens.append(tokens[i] + tokens[i+1])
            i += 2

        else:
            new_tokens.append(tokens[i])
            i += 1

    return new_tokens
```

```python
def get_vocab(tokens):

    vocab=set()

    for tok in tokens:
        vocab.add(tok)

    return vocab
```

```python
def train_bpe(text, max_vocab):
    tokens = ch_token(text)
    merges = []
    while True:
        pair_counts = count_pair(tokens)
        pair = best_pair(pair_counts)
        if pair is None:
            break
        if pair_counts[pair] == 1:
            break
        merges.append(pair)
        tokens = merge_pair(tokens, pair)
        vocab = get_vocab(tokens)
        print("merge:", pair, " vocab size:", len(vocab))
        if len(vocab) >= max_vocab:
            break
    return tokens, merges
```

```python
def build_vocab(vocab):

    vocab_dict = {}
    id = 0

    for token in vocab:
        vocab_dict[token] = id
        id += 1

    return vocab_dict
```

```python
def reverse_vocab(vocab_dict):

    rev = {}

    for token in vocab_dict:
        rev[vocab_dict[token]] = token

    return rev
```

```python
def encode(text, merges, vocab_dict):

    tokens = ch_token(text)

    for pair in merges:
        tokens = merge_pair(tokens, pair)

    ids = []

    for token in tokens:
        ids.append(vocab_dict[token])

    return ids
```

```python
def decode(ids, rev_vocab):

    tokens = []

    for i in ids:
        tokens.append(rev_vocab[i])

    text = ""

    for token in tokens:
        if token == "</w>":
            text += " "
        else:
            text += token

    return text.strip()
```

```python
vocab=int(input('enter number of vocab:'))
text_in=input('enter:')
merge=train_bpe(text_in,vocab)
print('bpe of given text is :',merge, sep='\n')
```
