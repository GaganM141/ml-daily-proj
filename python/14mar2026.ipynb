```python
#Armstrong
num=153
temp,temp2=num,num
count,sum1=0,0
while temp>0:
    count+=1
    temp//=10
for i in range(count):
    last=temp2%10
    sum1+=(last**count)
    temp2//=10
if sum1==num:
    print('number is Armstrong','\n sum:',sum1,'')
else:
    print("number isn't a Armstrong")
```
