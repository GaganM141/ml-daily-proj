# 2. Write a program where machine guesses the number.

```python
n=int(input('enter a number:'))
import random
i=0
j=19999
count=0
while True:
    num=random.randint(i,j)
    if count>1000:
        print('too many guesses')
        break
    if num==n:
        print('machine guessed successfully',num)
        break
    elif num>n:
        print('guessed number grater than actual number',num)
        
        j=num
    else:
        print('guessed number smaller than actual number',num)
        i=num
    count+=1
```

# 4. Write a program to ask for password until right password is provided.

```python
pwd='12@#'
while True:
    password=input('enter your password')
    if pwd==password:
        print('login successful')
        break
```

# 5. Write a program to verify user using password and if incorrect multiple times wait for few seconds. 

```python
import time
c=0
pwd='abc@123'
while True:
    password=input('enter your password:')
    if password==pwd:
        print('login successful')
        break
    else:
        c+=1
        print('incorrect password try again')
        if c==5:
            print('enterd wrong password multiple times please wait for 10s')
            time.sleep(10)
            c=0
```
