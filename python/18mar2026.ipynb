# 1. Write a program to remove duplicate values in a list.


```python
l=eval(input('enter a list:'))
res=[]
for i in l:
    if i not in res:
        res.append(i)
print(res)
```

# 2. Write a program to toggle a string


```python
s=input('enter a string:')
r=''
for i in s:
    if 'a'<=i<='z':
        r+=chr(ord(i)-32)
    elif 'A'<=i<='Z':
        r+=chr(ord(i)+32)
    else:
        r+=i
print('toggle of string',s,' is ',r)
```

# 3. Write a program to print concanants of a string.


```python
s=input('enter a string:')
for i in s:
    if ('a'<=i<='z' or 'A'<=i<='Z') and i not in 'AEIOUaeiou':
        print(i, end="")
```

```python
s1='aabadaddddcadsc'
count={}
sr=''
for i in s1:
    if i not in count:
        count[i]=1
    else:
        count[i]+=1
for i in count:
    sr+=i
    sr+=str(count[i])
print(sr)
```
