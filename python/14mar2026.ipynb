```python
str=input('enter a string:')
reverse_str=''
i=1
while i<=len(str):
    reverse_str+=str[-i]
    i+=1
print(reverse_str)
```

```python
i=101
a=0
while i<0:
    l=i%10
    a=(a*10)+l
    i//=10
if i==a:
    print('palindrome')
else:
    print('not')
# logic problem: i is updating in while loop
```

```python
ii=101
i=ii
a=0
while i>0:
    # by dividng a number by 10, the reminder will be the last digit
    last=i%10
    a=(a*10)+last
    i//=10
if ii==a:
    print('palindrome')
else:
    print('not palindrome')
```
