# write a program to print only conconants of the given string


```python
s=input('enter a string: ')
c=''
for i in s:
    if ('a'<=i<='z' or 'A'<=i<='Z') and i not in 'AEIOUaeiou':
        c+=i
print(c)
        
```

# write a program to count number of occurances of character in a string

```python
s=input('enter a string:')
c={}
for i in s:
    if i not in c:
        c[i]=1
    else:
        c[i]+=1
print("dict of char and there count: \n",c)
```

# write a program to count number of characters same in same position in 2 strings
## ex:
## s1='aaabcedaa'
## s2='aacbdeaaa'
## output=5

```python
s1=input('enter the first string:')
s2=input('enter the second string:')
out=0
if len(s1)>len(s2):
    for i in range(len(s2)):
        if s1[i]==s2[i]:
            out+=1
else:
    for i in range(len(s1)):
        if s1[i]==s2[i]:
            out+=1  
print('char in same index for both str: \n',out)
```

# toggle a string
## convert lower case char into upper case and visa versa

```python
s=input('enter a string: ')
toggle=''
for i in s:
    if 'a'<=i<='z':
        toggle+=chr(ord(i)-32)
    elif 'A'<=i<='Z':
        toggle+=chr(ord(i)+32)
    else:
        toggle+=i
print('toggle of input string is :',toggle)
        
```
