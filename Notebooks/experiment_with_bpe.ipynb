```python
def clean(text):
    texts=''
    for ch in text:
        if ch.isalnum() or ch==' ':
            texts+=ch.lower()
    return texts
```

```python
def ch_token(text):
    clean_text=clean(text)
    ch_text=[]
    for ch in clean_text:
        if ch!=' ':
            ch_text.append(ch)
        else:
            ch_text.append('</w>')
    #print('ch_text:',ch_text,sep='\n',end='\n\n\n')
    return ch_text
    
```

```python
def count_pair(text):
    ch_tokens=text
    Dcount_pair={}
    for i in range(0,len(ch_tokens)-1):
        if ch_tokens[i]=='</w>':
            continue
        pair=(ch_tokens[i],ch_tokens[i+1])
        if '</w>' not in pair:
            if pair not in Dcount_pair:
                Dcount_pair[pair]=1
            else:
                Dcount_pair[pair]+=1
    #print("dcount_pair",Dcount_pair,sep='\n')
    return Dcount_pair
```

```python
def best_pair(text):
    pair_count=text
    max_count=0
    best=None
    for pair in pair_count:
        #print('pair',pair,'   pair_value',pair_count[pair],sep='     ',end='\n')
        if pair_count[pair]>max_count:
            max_count=pair_count[pair]
            #print('max count    ',max_count)
            best=pair
    #print('best',best)
    return best
            
```

```python
def merge_best_pair(tokens,pair):
    new_tokens=[]
    i=0
    while i<len(tokens):
        if i<len(tokens)-1 and tokens[i]==pair[0] and tokens[i+1]==pair[1]:
            merged=tokens[i]+tokens[i+1]
            new_tokens.append(merged)
            i+=2
        else:
            new_tokens.append(tokens[i])
            i+=1
    return new_tokens
    
```

```python
def train_bpe(text,iterations):
    tokens=ch_token(text)
    for i in range(iterations):
        pair_counts=count_pair(tokens)
        pair=best_pair(pair_counts)
        if pair_counts[pair]==1 :
            break
        #print('merge pair:',pair)
        tokens=merge_best_pair(tokens,pair)
        #print('updated tokens:',tokens)
    return tokens
```

```python
text_in=input('enter')
bpe=train_bpe(text_in,1000)
print('bpe of given text is :',bpe, sep='\n')
```
