```python
def clean(text):
    texts=''
    for ch in text:
        if ch.isalnum() or ch==' ':
            texts+=ch.lower()
    return texts
```

```python
def ch_token(text):
    clean_text=clean(text)
    ch_text=[]
    for ch in clean_text:
        if ch!=' ':
            ch_text.append(ch)
        else:
            ch_text.append('</w>')
    return ch_text
    
```

```python
def count_pair(text):
    ch_tokens=ch_token(text)
    Dcount_pair={}
    for i in range(0,len(ch_tokens)-1):
        if ch_tokens[i]=='</w>':
            continue
        pair=(ch_tokens[i],ch_tokens[i+1])
        if '</w>' not in pair:
            if pair not in Dcount_pair:
                Dcount_pair[pair]=1
            else:
                Dcount_pair[pair]+=1
    return Dcount_pair
```

```python
def merge_frequent_pair(text):
    count_pairs=count_pair(text)
    pairs=list(count_pairs.keys())
    print(pairs)
    merge_ch={}
    for i in count_pairs:
        if count_pairs[i]!=1:
            #print(i,"    ",count_pairs[i])
            merge=''
            #print(pairs[i])
            for j in range(0,len(pairs[i])):
                merge+=pairs[i][j]
                print(j,"   ",merge)
            merge_ch[merge]=1
        else:
            continue
    return merge_ch
       
```

```python
text_in=input('enter the web text')
count=merge_frequent_pair(text_in)
print(count)
```
