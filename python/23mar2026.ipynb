# 1. Write a program to get 
### s='always keep  smiling'
### out='syawla peek gnilims'

```python
s=input('enter a string:')
s+=' '
reverse_word=''
out=''
for i in s:
    if i!=" ":
        reverse_word+=i
    else:
        j=len(reverse_word)
        while j>0:
            out+=reverse_word[j-1]
            j-=1
        reverse_word=''
    out+=' '
    print(' ')
print(out)
```

```python
s=input('enter a string:')
s+=" "
res=''
word=''
for i in s:
    if i!=" ":
        word+=i
    else:
        res+=word[::-1]+i
        word=''
print(res)
```

# 2. Write a program to check whether the number is prime number or not

```python
n=int(input('enter a number:'))
if n==1:
    print('not a prime number')
else:
    for i in range(2,n):
        if n%i==0:
            print('not a prime number')
            break
    else:
        print('prime number')
```
